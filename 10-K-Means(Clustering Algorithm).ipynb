# K-Means (Clustering Algorithm)

Note:-<br>
1)Clustering  is a part of partitioning datasets into groups called clusters.<br>
2)The goal is to split the data in such a way that points within a single cluster are very similar and points in different 
clusters are different.

K-Means CLustering<br>

K-means is one of the simple and most commonly used clustering algorithms.<br>
The algorithm alternates b/w two steps.<br>
 1- Assigning each data point to the closest cluster center.<br>
 2- And then setting each cluster center as the mean of the data points that are assign to it.<br>
Lets understand thoroughly

```python
import mglearn

mglearn.plots.plot_kmeans_algorithm()
```

```python
#By the given graphs you may understand that the data points are initialized and reassigned until it reaches the center.

#Next shows the boundaries of the cluster centers that are learned in the above figure
```

```python
mglearn.plots.plot_kmeans_boundaries()
```

```python
#Here, we apply k-means to the synthetic data that we used for preceding plots.
# 1-We instantiate the KMeans class, 2- and set the number of clusters we are looking for, 3- then we will call fit method.
#If you dont set up the number of clusters(n_clusters) then by default it is set to 8
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

#generate synthetic two-dimentional data
X, y = make_blobs(random_state=1)

#build the clustering model
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
#During the algorithm, each data point in X is assigned a cluster label.
#You can find these labels in the kmeans.label_attribute:
```

```python
print("Cluster memberships:\n{}".format(kmeans.labels_))
```

```python
#As we asked for three clusters, the clusters are numbered 0 to 2.

#You can also assign labels to new points, using the predict method.
#Each new point is assigned to the closest cluster center when predicting, but the existing model is not changed.
#Running predict on the training set returns the same result as labels_:
```

```python
print(kmeans.predict(X))
```

```python
#You can see that clustering is somewhat similar to classification, in that each item gets label
#Running the algorithm again might result in a different numbering of clusters because of the random nature of initialization
#Here is a plot of this data again.
#The cluster centers are stored in the cluster_centers_attribute, and we plot them as triangles.
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], kmeans.labels_, markers='o')
mglearn.discrete_scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], [0,1,2],
                         markers='^', markeredgewidth=2)
```

```python
#We can also use more or few cluster centers.
```

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(13,7))

#using two cluster centers
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
assignments = kmeans.labels_

mglearn.discrete_scatter(X[:, 0], X[:, 1], assignments, ax=axes[0])

#using five cluster centers
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
assignments = kmeans.labels_

mglearn.discrete_scatter(X[:, 0], X[:, 1], assignments, ax=axes[1])
```
